## 3. Analysis

### 3.1 Proper Visualization of the Dataset

```python
#Category vs Sub-Category

plt.figure(figsize=(20,8))
plt.bar('Sub-Category','Category', data=sample_store)
sns.set(style="darkgrid")
plt.title('Category vs Sub-Category', fontsize=20)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Category', fontsize=15)
plt.show()
```

```python
#Print the state count

print(ss['State'].value_counts())
```

```python
#Count Plot for State

plt.figure(figsize=(15,8))
sns.countplot(x=ss['State'])
plt.xlabel('State', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#Print the Sub-Category

print(ss['Sub-Category'].value_counts())
```

```python
#Count Plot for Sub-Category

plt.figure(figsize=(12,6))
sns.countplot(x=sample_store['Sub-Category'])
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

### 3.2 Heatmap of Dataset

```python
#Heatmap of coorelation

fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(sample_store.corr(), annot= True, cmap='Spectral')
plt.show()
```

```python
#Heatmap of Covariance

fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(sample_store.cov(), annot= True, cmap= 'Spectral')
plt.show()
```

```python
#Count Plot for Segment

plt.figure(figsize=[6,6])
sns.countplot(x=sample_store['Segment'])
plt.xlabel('Segment', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

```python
#Count Plot for Region

plt.figure(figsize=[8,6])
sns.countplot(x=sample_store['Region'])
plt.xlabel('Region', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

```python
#Bar Plot for Sub-Category vs Profit

plt.figure(figsize=(20,12))
sns.barplot(x=sample_store['Sub-Category'], y=sample_store['Profit'])
plt.title('Sub-Category vs Profit', fontsize=20)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.show()
```

```python
#line Graph for Profit vs Discount

plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = sample_store, color = 'r', label= 'Discount')
plt.xlabel('Discount', fontsize=15)
plt.ylabel('Profit', fontsize=15)
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

```python
#Histogram for Sub-Category

ss.hist(bins=50 ,figsize=(20,15))
plt.show()
```

```python
#Pair Plot for Sub-Category

figsize=(15,10)
sns.pairplot(data=ss, hue='Sub-Category')
plt.show()
```

### 3.4 So Now we Grouped or sum the sales, profit, discount, quantity according to every state of region and also according to sub-categories sales

```python
grouped = pd.DataFrame(sample_store.groupby(['Ship Mode','Segment','Category','Sub-Category','State','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped
```

```python
#sum, mean, min, max, count, median, standard deviation, Variance of each states of Profit

sample_store.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```
